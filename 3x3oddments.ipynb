Oddments

```python
import numpy
def oddments(p1gametable,outputvar):
    try:
        outputvar.set("")
        #P1 oddments
        c1c2 = []
        c2c3 = []

        p1oddments = []

        for x in range(3):
            c1c2.append(p1gametable[x][0]-p1gametable[x][1])
            c2c3.append(p1gametable[x][1]-p1gametable[x][2])
        
        totaloddments = 0

        for x in range(3):
            matrix = numpy.zeros((2,2))
            index = 0
            for y in range(3):
                if(y != x):
                    matrix[index][0] = c1c2[y]
                    matrix [index][1]= c2c3[y]
                    index += 1
            #calculate oddment
            det = numpy.linalg.det(matrix)
            p1oddments.append(abs(det))
            totaloddments += abs(det)
        if(totaloddments == 0):
            outputvar.set("Something went wrong. P1's oddments summed to 0")
        outputvar.set(f"Player 1 should play strategy 0 with probability {p1oddments[0]/totaloddments}, strategy 1 with probability {p1oddments[1]/totaloddments}, and strategy 2 with probability {p1oddments[2]/totaloddments} \n")


        #P2oddments
        r1r2 = []
        r2r3 = []

        p2oddments = []

        for x in range(3):
            r1r2.append(p1gametable[0][x]-p1gametable[1][x])
            r2r3.append(p1gametable[1][x]-p1gametable[2][x])
        
        totaloddments = 0

        for x in range(3):
            matrix = numpy.zeros((2,2))
            index = 0
            for y in range(3):
                if(y != x):
                    matrix[0][index] = r1r2[y]
                    matrix[1][index] = r2r3[y]
                    index += 1
            #calculate oddment
            det = numpy.linalg.det(matrix)
            p2oddments.append(abs(det))
            totaloddments += abs(det)
        if(totaloddments == 0):
            outputvar.set("Something went wrong. P2's oddments sum  to 0")
        outputvar.set(outputvar.get() + f"Player 2 should play strategy 0 with probability {p2oddments[0]/totaloddments}, strategy 1 with probability {p2oddments[1]/totaloddments}, and strategy 2 with probability {p2oddments[2]/totaloddments}")
    except(ZeroDivisionError):
        outputvar.set("Something went wrong. One of the oddments summed to 0")


def Equilibreum(gametableforp1txt, outputvar):
    gametableforp1 = []
    gametableforp2 = []

    try:
        for x in range(3):
            curr1 = []
            curr2 = []
            for y in range(3):
                curr1.append(float(gametableforp1txt[x][y].get()))
                curr2.append(-float(gametableforp1txt[x][y].get()))
            gametableforp1.append(curr1)
            gametableforp2.append(curr2)
    except(ValueError):
        outputvar.set("We couldn't read one of your numbers. Please check to see that there are no letters or spaces")
    
    dimentions = 3
    anyequilibreum = False
    output = ''
    for x in range(dimentions):
        for y in range(dimentions):
            #Check for better strat for P2
            p2payoff = gametableforp2[x][y]
            equilibreum = True
            for z in range(dimentions):
                if(z != y):
                    if(gametableforp2[x][z] > p2payoff):
                        equilibreum = False
            
            #Check for better strat for P1
            p1payoff = gametableforp1[x][y]
            for z in range(dimentions):
                if(z != x):
                    if(gametableforp1[z][y] > p1payoff):
                        equilibreum = False
            if(equilibreum):
                anyequilibreum = True
                output += f"There is an equilibreum at {x},{y}\n"
    if(anyequilibreum == False):
        output = "Run RepeatedStrategies"
    outputvar.set(output)
    if(output == "Run RepeatedStrategies"):
        oddments(gametableforp1,outputvar)


 
```

Tkinter Window

```python
import tkinter
root = tkinter.Tk()
frame = tkinter.Frame(root)
frame.pack()
p1textvars = []
outputvar = tkinter.StringVar(frame)
for x in range(3):
    curr = []
    tkinter.Label(frame,text=str(x)).grid(row = x+1, column = 0)
    for y in range(3):
        curr.append(tkinter.StringVar(frame, value= f"p1 value for [{x},{y}]"))
        tkinter.Entry(frame, textvariable=curr[y]).grid(row = x+1,column= (y)+1)
        if(x == 0):
            tkinter.Label(frame,text = str(y)).grid(row=0,column=y+1)
    p1textvars.append(curr)
tkinter.Button(frame,text="Find the Equilibreum", command=lambda:Equilibreum(p1textvars,outputvar)).grid(row = 8, column = 2)
#tkinter.Button(frame,text="Find the Movement Table", command=lambda:FindMovementTable(p1textvars,outputvar)).grid(row = 8, column = 1)
tkinter.Label(frame,textvariable=outputvar).grid(row=10,column = 0, columnspan=9)
root.mainloop()
```
